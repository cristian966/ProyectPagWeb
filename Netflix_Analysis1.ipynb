```python
#Acceder a columnas y contar valores
#Atributo de normalize nos permite mostrar el conteo en porcentaje
#Para ordenar
paises = netflix["country"].value_counts(normalize = True, ascending = False)
print(paises)
```

```python
#Analizar columnas con el metodo describe()
descripcion_directores = netflix["director"].describe()
print(descripcion_directores)
```

```python
#ahora con numeros
descripcion_lanzamiento = netflix["release_year"].describe()
print(descripcion_lanzamiento)
```

```python
#Ordenar por valores de una columna
year_sort = netflix.sort_values(by = "release_year", ascending = True)
year_sort 
```

```python
#Ordenar por indice(id)
year_sort_index = netflix.sort_index(ascending = False)
year_sort_index
```
